```python
# 必要なライブラリをインポート
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
import warnings
warnings.filterwarnings('ignore')  # 警告を非表示

# 検索オプションのためのウィジェットを作成
keyword_input = widgets.Text(
    value='機械学習',
    placeholder='検索キーワードを入力',
    description='キーワード:',
    disabled=False
)

timeframe_dropdown = widgets.Dropdown(
    options=[
        ('過去1年', 'today 12-m'),
        ('過去3ヶ月', 'today 3-m'),
        ('過去1ヶ月', 'today 1-m'),
        ('過去5年', 'today 5-y'),
        ('2004年以降すべて', 'all')
    ],
    value='today 12-m',
    description='期間:',
    disabled=False,
)

geo_dropdown = widgets.Dropdown(
    options=[
        ('日本', 'JP'),
        ('アメリカ', 'US'),
        ('イギリス', 'GB'),
        ('全世界', ''),
        ('中国', 'CN'),
        ('韓国', 'KR')
    ],
    value='JP',
    description='地域:',
    disabled=False,
)

property_dropdown = widgets.Dropdown(
    options=[
        ('Web検索', ''),
        ('YouTube検索', 'youtube'),
        ('ニュース検索', 'news'),
        ('画像検索', 'images'),
        ('ショッピング検索', 'froogle')
    ],
    value='',
    description='検索プロパティ:',
    disabled=False,
)

compare_checkbox = widgets.Checkbox(
    value=False,
    description='2つ目のキーワードを追加',
    disabled=False
)

keyword2_input = widgets.Text(
    value='AI',
    placeholder='2つ目のキーワードを入力',
    description='キーワード2:',
    disabled=False
)

compare_properties_checkbox = widgets.Checkbox(
    value=False,
    description='Web検索とYouTube検索を比較',
    disabled=False
)

# 実行ボタン
run_button = widgets.Button(
    description='分析を実行',
    disabled=False,
    button_style='success',
    tooltip='クリックして分析を実行',
    icon='check'
)

# ウィジェットの更新用関数
def update_widgets(change):
    keyword2_input.disabled = not compare_checkbox.value

compare_checkbox.observe(update_widgets, names='value')

# メイン関数
def run_analysis(b):
    clear_output(wait=True)

    # ウィジェットを再表示
    display(keyword_input, timeframe_dropdown, geo_dropdown, property_dropdown,
            compare_checkbox, keyword2_input, compare_properties_checkbox, run_button)

    # 入力を取得
    keyword = keyword_input.value
    timeframe = timeframe_dropdown.value
    geo = geo_dropdown.value
    prop = property_dropdown.value
    compare = compare_checkbox.value
    keyword2 = keyword2_input.value if compare else None
    compare_props = compare_properties_checkbox.value

    print(f"分析を実行中: {keyword}" + (f" vs {keyword2}" if compare else ""))
    print(f"期間: {timeframe_dropdown.options[timeframe_dropdown.index][0]}, 地域: {geo_dropdown.options[geo_dropdown.index][0]}")

    # pytrendsの分析実行
    try:
        # pytrends APIの初期化
        pytrends = TrendReq(hl='ja-JP', tz=360)

        if compare:
            # 2つのキーワードを比較
            kw_list = [keyword, keyword2]

            if compare_props:
                # WebとYouTubeの両方で検索
                print("\n通常のウェブ検索トレンドを取得中...")
                pytrends.build_payload(kw_list, cat=0, timeframe=timeframe, geo=geo, gprop='')
                interest_over_time_web = pytrends.interest_over_time()

                print("YouTube検索トレンドを取得中...")
                pytrends.build_payload(kw_list, cat=0, timeframe=timeframe, geo=geo, gprop='youtube')
                interest_over_time_youtube = pytrends.interest_over_time()

                # データフレームを結合
                if not interest_over_time_web.empty and not interest_over_time_youtube.empty:
                    # カラム名を変更
                    interest_over_time_web = interest_over_time_web.rename(columns={
                        keyword: f"{keyword}_web",
                        keyword2: f"{keyword2}_web"
                    })

                    interest_over_time_youtube = interest_over_time_youtube.rename(columns={
                        keyword: f"{keyword}_youtube",
                        keyword2: f"{keyword2}_youtube"
                    })

                    # isPartial列を削除
                    if 'isPartial' in interest_over_time_web.columns:
                        interest_over_time_web = interest_over_time_web.drop('isPartial', axis=1)
                    if 'isPartial' in interest_over_time_youtube.columns:
                        interest_over_time_youtube = interest_over_time_youtube.drop('isPartial', axis=1)

                    # データフレームを結合
                    combined_df = pd.concat([interest_over_time_web, interest_over_time_youtube], axis=1)

                    # 相関行列の計算
                    correlation_matrix = combined_df.corr()
                    print("\n相関行列:")
                    print(correlation_matrix)

                    # ヒートマップで相関行列を可視化
                    plt.figure(figsize=(10, 8))
                    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
                    plt.title('Correlation Matrix of Web and YouTube Search Trends', fontsize=16)
                    plt.tight_layout()
                    plt.show()

                    # 時系列プロット
                    plt.figure(figsize=(14, 7))

                    # Web検索のプロット
                    plt.plot(combined_df.index, combined_df[f"{keyword}_web"],
                             color='blue', label=f'Web: {keyword}')
                    plt.plot(combined_df.index, combined_df[f"{keyword2}_web"],
                             color='green', label=f'Web: {keyword2}')

                    # YouTube検索のプロット - 点線で区別
                    plt.plot(combined_df.index, combined_df[f"{keyword}_youtube"],
                             color='blue', linestyle='--', label=f'YouTube: {keyword}')
                    plt.plot(combined_df.index, combined_df[f"{keyword2}_youtube"],
                             color='green', linestyle='--', label=f'YouTube: {keyword2}')

                    # グラフの設定
                    plt.title(f'Comparison of Search Trends: "{keyword}" vs "{keyword2}"', fontsize=16)
                    plt.xlabel('Date')
                    plt.ylabel('Interest (0-100)')
                    plt.legend()
                    plt.grid(True, linestyle='--', alpha=0.7)
                    plt.tight_layout()
                    plt.show()

                    # データをCSVファイルに保存
                    csv_filename = f'{keyword}_vs_{keyword2}_web_youtube.csv'
                    combined_df.to_csv(csv_filename)
                    print(f"\nデータを '{csv_filename}' として保存しました。")
                else:
                    print("データが取得できませんでした。別のキーワードや期間を試してください。")
            else:
                # 選択したプロパティのみで検索
                pytrends.build_payload(kw_list, cat=0, timeframe=timeframe, geo=geo, gprop=prop)
                interest_over_time_df = pytrends.interest_over_time()

                if not interest_over_time_df.empty:
                    # 基本統計量の表示
                    print("\n基本的な統計量:")
                    print(interest_over_time_df.describe())

                    # 相関係数の計算
                    correlation = interest_over_time_df[keyword].corr(interest_over_time_df[keyword2])
                    print(f"\n「{keyword}」と「{keyword2}」の相関係数: {correlation:.4f}")

                    # 相関の強さを解釈
                    if abs(correlation) < 0.3:
                        strength = "弱い"
                    elif abs(correlation) < 0.7:
                        strength = "中程度の"
                    else:
                        strength = "強い"

                    direction = "正の" if correlation >= 0 else "負の"
                    print(f"これは{strength}{direction}相関を示しています。")

                    # 時系列グラフ
                    plt.figure(figsize=(14, 7))
                    plt.plot(interest_over_time_df.index, interest_over_time_df[keyword], color='blue', label=keyword)
                    plt.plot(interest_over_time_df.index, interest_over_time_df[keyword2], color='red', label=keyword2)

                    prop_name = property_dropdown.options[property_dropdown.index][0]
                    plt.title(f'Comparison of {prop_name} Trends: "{keyword}" vs "{keyword2}"', fontsize=16)
                    plt.xlabel('Date')
                    plt.ylabel('Interest (0-100)')
                    plt.legend()
                    plt.grid(True, linestyle='--', alpha=0.7)
                    plt.tight_layout()
                    plt.show()

                    # 散布図と回帰線
                    plt.figure(figsize=(10, 6))
                    sns.regplot(x=interest_over_time_df[keyword], y=interest_over_time_df[keyword2],
                                scatter_kws={"alpha":0.5}, line_kws={"color":"red"})

                    plt.title(f'Correlation between "{keyword}" and "{keyword2}" (r={correlation:.4f})', fontsize=16)
                    plt.xlabel(f'Interest in "{keyword}" (0-100)')
                    plt.ylabel(f'Interest in "{keyword2}" (0-100)')
                    plt.grid(True, linestyle='--', alpha=0.3)
                    plt.tight_layout()
                    plt.show()

                    # データをCSVファイルに保存
                    prop_text = "" if prop == "" else f"_{prop}"
                    csv_filename = f'{keyword}_vs_{keyword2}{prop_text}.csv'
                    interest_over_time_df.to_csv(csv_filename)
                    print(f"\nデータを '{csv_filename}' として保存しました。")
                else:
                    print("データが取得できませんでした。別のキーワードや期間を試してください。")
        else:
            # 単一キーワードの分析
            kw_list = [keyword]

            if compare_props:
                # WebとYouTubeの両方で検索
                print("\n通常のウェブ検索トレンドを取得中...")
                pytrends.build_payload(kw_list, cat=0, timeframe=timeframe, geo=geo, gprop='')
                interest_over_time_web = pytrends.interest_over_time()

                print("YouTube検索トレンドを取得中...")
                pytrends.build_payload(kw_list, cat=0, timeframe=timeframe, geo=geo, gprop='youtube')
                interest_over_time_youtube = pytrends.interest_over_time()

                # データフレームを結合
                if not interest_over_time_web.empty and not interest_over_time_youtube.empty:
                    # カラム名を変更
                    interest_over_time_web = interest_over_time_web.rename(columns={keyword: f"{keyword}_web"})
                    interest_over_time_youtube = interest_over_time_youtube.rename(columns={keyword: f"{keyword}_youtube"})

                    # isPartial列を削除
                    if 'isPartial' in interest_over_time_web.columns:
                        interest_over_time_web = interest_over_time_web.drop('isPartial', axis=1)
                    if 'isPartial' in interest_over_time_youtube.columns:
                        interest_over_time_youtube = interest_over_time_youtube.drop('isPartial', axis=1)

                    # データフレームを結合
                    combined_df = pd.concat([interest_over_time_web, interest_over_time_youtube], axis=1)

                    # 相関係数を計算
                    correlation = combined_df[f"{keyword}_web"].corr(combined_df[f"{keyword}_youtube"])
                    print(f"\nWeb検索とYouTube検索の相関係数: {correlation:.4f}")

                    # 相関の強さを解釈
                    if abs(correlation) < 0.3:
                        strength = "弱い"
                    elif abs(correlation) < 0.7:
                        strength = "中程度の"
                    else:
                        strength = "強い"

                    direction = "正の" if correlation >= 0 else "負の"
                    print(f"これは{strength}{direction}相関を示しています。")

                    # データの可視化
                    plt.figure(figsize=(14, 7))
                    plt.plot(combined_df.index, combined_df[f"{keyword}_web"],
                             color='blue', label=f'Web Search')
                    plt.plot(combined_df.index, combined_df[f"{keyword}_youtube"],
                             color='red', label=f'YouTube Search')

                    plt.title(f'Comparison of Web and YouTube Search Trends for "{keyword}"', fontsize=16)
                    plt.xlabel('Date')
                    plt.ylabel('Interest (0-100)')
                    plt.legend()
                    plt.grid(True, linestyle='--', alpha=0.7)
                    plt.tight_layout()
                    plt.show()

                    # 散布図と回帰線
                    plt.figure(figsize=(10, 6))
                    sns.regplot(x=combined_df[f"{keyword}_web"], y=combined_df[f"{keyword}_youtube"],
                                scatter_kws={"alpha":0.5}, line_kws={"color":"red"})

                    plt.title(f'Correlation between Web and YouTube Search for "{keyword}" (r={correlation:.4f})', fontsize=16)
                    plt.xlabel(f'Web Search Interest (0-100)')
                    plt.ylabel(f'YouTube Search Interest (0-100)')
                    plt.grid(True, linestyle='--', alpha=0.3)
                    plt.tight_layout()
                    plt.show()

                    # データをCSVファイルに保存
                    csv_filename = f'{keyword}_web_youtube_comparison.csv'
                    combined_df.to_csv(csv_filename)
                    print(f"\nデータを '{csv_filename}' として保存しました。")
                else:
                    print("データが取得できませんでした。別のキーワードや期間を試してください。")
            else:
                # 選択したプロパティのみで検索
                pytrends.build_payload(kw_list, cat=0, timeframe=timeframe, geo=geo, gprop=prop)
                interest_over_time_df = pytrends.interest_over_time()

                if not interest_over_time_df.empty:
                    # 取得したデータの確認
                    print("\nデータの最初の5行:")
                    print(interest_over_time_df.head())

                    # 基本的な統計量を表示
                    print("\n基本的な統計量:")
                    print(interest_over_time_df.describe())

                    # データの可視化
                    plt.figure(figsize=(12, 6))
                    plt.plot(interest_over_time_df.index, interest_over_time_df[keyword], color='blue')

                    prop_name = property_dropdown.options[property_dropdown.index][0]
                    plt.title(f'{prop_name} Trend for "{keyword}"', fontsize=16)
                    plt.xlabel('Date')
                    plt.ylabel('Interest (0-100)')
                    plt.grid(True, linestyle='--', alpha=0.7)
                    plt.tight_layout()
                    plt.show()

                    # データをCSVファイルに保存
                    prop_text = "" if prop == "" else f"_{prop}"
                    csv_filename = f'{keyword}{prop_text}_trend_data.csv'
                    interest_over_time_df.to_csv(csv_filename)
                    print(f"\nデータを '{csv_filename}' として保存しました。")
                else:
                    print("データが取得できませんでした。別のキーワードや期間を試してください。")
    except Exception as e:
        print(f"エラーが発生しました: {e}")
        print("別のキーワードや期間を試してください。")

# ボタンクリック時の処理を設定
run_button.on_click(run_analysis)

# 初期状態ではキーワード2を無効化
keyword2_input.disabled = True

# ウィジェットを表示
display(keyword_input, timeframe_dropdown, geo_dropdown, property_dropdown,
        compare_checkbox, keyword2_input, compare_properties_checkbox, run_button)
```
